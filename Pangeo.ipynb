# Grab files we want to "download" from Pangeo

```python
import intake
import xarray as xr
import fsspec
```

```python
# See what's available for the variable, experiment, grid(s), monthly output:
```

```python
cat_url = "https://cmip6.storage.googleapis.com/pangeo-cmip6.json"
col = intake.open_esm_datastore(cat_url)
cat = col.search(experiment_id=['historical'], variable_id='fgco2',
                 grid_label='gr', table_id='Omon')
dset_dict = {}
for name, group in cat.df.groupby('source_id'):
    dsets = [xr.open_zarr(fsspec.get_mapper(zstore), consolidated=True)
             for zstore in group.zstore.values]
# ADD BIT BELOW TO LOAD THE DATA ONTO NOTEBOOK
    try:
        dsets_aligned = xr.align(*dsets, join='exact')
        dset_dict[name] = xr.concat(dsets, dim='member')
        print(f'concatted {name}')
#        ds = dset_dict[f"{name}"]
#        ds.to_netcdf(f"/data/artemis/simulations/CMIP6/RCP26/{name}_fgco2.nc")
    except Exception as e:
        print(f'failed to concat {name}')
        print(e)
```
